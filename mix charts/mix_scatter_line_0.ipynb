# Libraries

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
```

### Activate latex font format

Do you want plot with latex font format? Use the cell bellow:

```python
mpl.rcParams.update({
                        'font.family': 'serif',
                        'mathtext.fontset': 'cm',
                        'axes.unicode_minus': False
                    })
```

# Dataset upload or create

For simplicity we use a random variable concept to create a dataset.

```python
n = 200
observed = np.linspace(0.4, 2.5, n)
multiplicative_noise = np.random.lognormal(mean=0.0, sigma=0.08, size=n)
additive_noise = np.random.normal(loc=0.0, scale=0.03, size=n)
predicted = 0.92 * observed * multiplicative_noise + additive_noise
predicted = np.clip(predicted, a_min=0.0, a_max=None)
data = pd.DataFrame({
                        "Observed": observed,
                        "Predicted": predicted
                    })
```

# Chart

### Figure name and DPI

```python
dpi = 600               # Change as you wish
name = 'sine_wave_plot' # Change as you wish
```

### Size in centimeters

```python
b_cm = 12                       # Change as you wish
h_cm = 8                        # Change as you wish
inches_to_cm = 1 / 2.54
b_input = b_cm * inches_to_cm
h_input = h_cm * inches_to_cm
```

### Axis, Labels, Legends and general properties

$\$\;\text{math expression}\;\$$ represents math-latex symbol

```python
label_x = 'Observed'            # Change as you wish
label_y = 'Predicted'           # Change as you wish
size_label = 14                 # Change as you wish
color_label = 'blue'            # or hexadecimal. Change as you wish
size_axis = 10                  # Change as you wish
color_axis = 'red'              # or hexadecimal. Change as you wish
```

### Lines

```python
size_line = 4           # Change as you wish
style_line = '--'       # Change as you wish
color_line = 'red'      # or hexadecimal. Change as you wish
```

### Scatter

```python
alpha_scatter = 0.4         # Change as you wish
color_scatter = 'blue'      # Change as you wish
size_scatter = 25           # Change as you wish
```

### Legend

```python
labels_legend = [r'$R^2$=0.98']   # Change as you wish
size_legend = 10                # Change as you wish
location_legend = 'upper left'  # Change as you wish - 'best' look up by the best fit
```

### Figure

```python
fig, ax = plt.subplots(figsize=(b_input, h_input))
lims = [
            min(data.min()),
            max(data.max())
        ]   
"""
# Title. Do you need a title? Use the cell bellow:

ax.set_title('Sine Wave Plot', fontsize=16)
"""
# Config axis
ax.tick_params(axis='both', which='major', labelsize=size_axis, colors=color_axis)
ax.set_xlabel(label_x, fontsize=size_label, color=color_label)
ax.set_ylabel(label_y, fontsize=size_label, color=color_label)

# Config grid
on_or_off = True
plt.grid(on_or_off, which='both', linestyle='-', linewidth=0.5)

# Plot data y = x and scatter
ax.plot(lims, lims, linewidth=size_line, linestyle=style_line, color=color_line,)
ax.scatter(data["Observed"], data["Predicted"], alpha=alpha_scatter, color=color_scatter, s=size_scatter, label=labels_legend[0])
ax.legend(fontsize=size_legend, loc=location_legend)

# Save. Do you need save? Use the cell bellow:
fig.savefig('Traditional-mix-scatter-line.png', dpi=dpi, bbox_inches='tight')
plt.show()
```
