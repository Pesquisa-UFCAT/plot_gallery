# Libraries

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
```

### Activate latex font format

Do you want plot with latex font format? Use the cell bellow:

```python
mpl.rcParams.update({
    'font.family': 'serif',
    'mathtext.fontset': 'cm',
    'axes.unicode_minus': False
})
```

# Dataset upload or create

For simplicity we use a random variable concept to create a dataset.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2
```

# Chart

### Figure name and DPI

```python
dpi = 600               # Change as you wish
name = 'sine_wave_plot' # Change as you wish
```

### Size in centimeters

```python
b_cm = 12                     # Change as you wish
h_cm = 8                      # Change as you wish
inches_to_cm = 1 / 2.54
b_input = b_cm * inches_to_cm
h_input = h_cm * inches_to_cm
```

### Axis, Labels, Legends and general properties

$\$\;\text{math expression}\;\$$ represents math-latex symbol

```python
label_x = '$x$'      # Change as you wish
label_y = '$y$'      # Change as you wish
size_label = 14      # Change as you wish
color_label = 'blue' # or hexadecimal. Change as you wish
size_axis = 12       # Change as you wish
color_axis = 'cyan'   # or hexadecimal. Change as you wish
```

### Lines

```python
size_line = 2        # Change as you wish
style_line = '-'     # Change as you wish
color_line = 'green' # or hexadecimal. Change as you wish
```

### Contour levels

```python
cmap_type = 'Reds'   # Change as you wish. Check matplotlib colormaps documentation: https://matplotlib.org/stable/users/explain/colors/colormaps.html
alpha_level = 1      # From 0 to 1. Change as you wish
levels = 10          # Change as you wish
```

### Figure

```python
fig, ax = plt.subplots(figsize=(b_input, h_input))
"""
# Title. Do you need a title? Use the cell bellow:

ax.set_title('Sine Wave Plot', fontsize=16)
"""
# Config axis
ax.tick_params(axis='both', which='major', labelsize=size_axis, colors=color_axis)
ax.set_xlabel(label_x, fontsize=size_label, color=color_label)
ax.set_ylabel(label_y, fontsize=size_label, color=color_label)

# Plot data
liness = ax.contourf(x, y, z, levels=levels, cmap=cmap_type, alpha=alpha_level)
cbar = fig.colorbar(liness, ax=ax)
cbar.ax.tick_params(labelsize=size_axis)

# Save. Do you need save? Use the cell bellow:

fig.savefig('Traditional-contourplot-With-Cmap.png', dpi=dpi, bbox_inches='tight')
plt.show()
```
