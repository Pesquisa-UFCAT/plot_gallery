# Libraries

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
```

### Activate latex font format

Do you want plot with latex font format? Use the cell bellow:

```python
mpl.rcParams.update({
    'font.family': 'serif',
    'mathtext.fontset': 'cm',
    'axes.unicode_minus': False
})
```

# Dataset upload or create

For simplicity we use a random variable concept to create a dataset.

```python
x = np.linspace(0, 4 * np.pi, 50)
y = np.sin(x)
m = np.linspace(0, 4 * np.pi, 50)
n = np.cos(m)
data = {
            'x0': x,
            'y0': y,
            'x1': m,
            'y1': n
        }
data = pd.DataFrame(data)
```

# Chart

### Figure name and DPI

```python
dpi = 600               # Change as you wish
name = 'sine_wave_plot' # Change as you wish
```

### Size in centimeters

```python
b_cm = 12                       # Change as you wish
h_cm = 8                        # Change as you wish
inches_to_cm = 1 / 2.54
b_input = b_cm * inches_to_cm
h_input = h_cm * inches_to_cm
```

### Axis, Labels, Legends and general properties

$\$\;\text{math expression}\;\$$ represents math-latex symbol

```python
label_x = '$\\theta$ (radians)' # Change as you wish
label_y = 'sin ($\\theta$)'     # Change as you wish
size_label = 14                 # Change as you wish
color_label = 'blue'            # or hexadecimal. Change as you wish
size_axis = 10                  # Change as you wish
color_axis = 'red'              # or hexadecimal. Change as you wish
```

### Lines

```python
size_line = 2                   # Change as you wish
style_line = ['--', '-.']        # Change as you wish
color_line = ['green', 'blue']  # or hexadecimal. Change as you wish
```

### Markers

```python
marker = None                    # Change as you wish
size_marker = 4                 # Change as you wish
color_marker = 'orange'         # or hexadecimal. Change as you wish
color_marker_edge = color_line  # Change as you wish
```

### Legend

```python
labels_legend = ['Sine Wave', 'Cosine Wave']   # Change as you wish
size_legend = 10                # Change as you wish
location_legend = 'lower left'  # Change as you wish - 'best' look up by the best fit
```

### Figure

```python
fig, ax = plt.subplots(figsize=(b_input, h_input))
"""
# Title. Do you need a title? Use the cell bellow:

ax.set_title('Sine Wave Plot', fontsize=16)
"""
# Config axis
ax.tick_params(axis='both', which='major', labelsize=size_axis, colors=color_axis)
ax.set_xlabel(label_x, fontsize=size_label, color=color_label)
ax.set_ylabel(label_y, fontsize=size_label, color=color_label)

# Control axis. Do you need control the axis length? Use the cells below:

# Automatic calculation of axis limits
x_min = min(data['x0'])
x_max = max(data['x0'])
y_min = min(data['y0'])
y_max = max(data['y0'])
margin_x = 0           # Your desired margin
margin_y = 0           # Your desired margin
ax.set_ylim(bottom=y_min - margin_y, top=y_max + margin_y)
ax.set_xlim(left=x_min - margin_x, right=x_max + margin_x)

""" or use manual limits:
# Your plotting code
leftt = 0.1
rightt = 0.1
bottomt = 0.1
topt = 0.1
ax.set_ylim(bottom=bottomt, top=topt)
ax.set_xlim(left=leftt, right=rightt)
"""

# Config grid
on_or_off = True
plt.grid(on_or_off, which='both', linestyle='-', linewidth=0.5)

# Plot data
ax.plot(data['x0'], data['y0'], 
         markerfacecolor=color_marker, marker=marker, markersize=size_marker,
         linewidth=size_line, linestyle=style_line[0], color=color_line[0],
         label=labels_legend[0])
ax.plot(data['x1'], data['y1'], 
         markerfacecolor=color_marker, marker=marker, markersize=size_marker,
         linewidth=size_line, linestyle=style_line[1], color=color_line[1],
         label=labels_legend[1])
ax.invert_xaxis()
ax.legend(fontsize=size_legend, loc=location_legend)

# Save. Do you need save? Use the cell bellow:
fig.savefig('Traditional-lineplot-invertaxis-2lines.png', dpi=dpi, bbox_inches='tight')
plt.show()
```
